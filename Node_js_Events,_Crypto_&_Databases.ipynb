# Question 1: Create a custom event emitter that fires a 'userLoggedIn' event with a username payload. Set up a listener to print a welcome message when the event occurs.


Ans 1:

Custom Event Emitter — <code>userLoggedIn</code> Event


Concept Overview


Node.js provides the built-in <code>EventEmitter</code> class from the <code>events</code> module. We can extend it to create custom emitters that fire named events with data payloads, and attach listeners that respond to those events.


Full Implementation

<pre>
// Import the built-in EventEmitter class
const EventEmitter = require('events');

// Step 1: Create a custom class that extends EventEmitter
class UserAuthEmitter extends EventEmitter {}

// Step 2: Instantiate the custom emitter
const authEmitter = new UserAuthEmitter();

// Step 3: Register a listener for the 'userLoggedIn' event
authEmitter.on('userLoggedIn', (payload) => {
  console.log(`Welcome, ${payload.username}! You have successfully logged in.`);
  console.log(`Login Time: ${payload.loginTime}`);
});

// Step 4: Emit the 'userLoggedIn' event with a username payload
authEmitter.emit('userLoggedIn', {
  username: 'Alice',
  loginTime: new Date().toLocaleString()
});
```

---

### Output
```
Welcome, Alice! You have successfully logged in.
Login Time: 12/3/2026, 10:45:00 AM
</pre>

Step-by-Step Breakdown

1. Import <code>EventEmitter</code>

The <code>events</code> module is a core Node.js module — no installation needed. <code>EventEmitter</code> is the base class for all event-driven behavior.

2. Extend <code>EventEmitter</code>

By extending <code>EventEmitter</code>, <code>UserAuthEmitter</code> inherits all methods like <code>.on()</code>, <code>.emit()</code>, <code>.once()</code>, and <code>.off()</code>.

3. Attach a Listener with <code>.on()</code>



- <code>.on(eventName, callback)</code> registers a persistent listener — it fires every time the event is emitted.

- The <code>payload</code> parameter receives the data object passed during <code>.emit()</code>.

4. Fire the Event with <code>.emit()</code>



<code>.emit(eventName, data)</code> triggers all listeners registered for <code>'userLoggedIn'</code>.
The second argument is the payload — any data you want to send with the event.

Why Use Custom Event Emitters?



- Decoupling: The login logic doesn't need to know who is listening.
- Scalability: Multiple parts of an app can react to the same event independently.
- Reusability: The emitter pattern is the foundation of Node.js streams, HTTP - servers, and more.

# Question 2: Use the crypto module to create an SHA256 hash for the string "password123".


Ans 2:
SHA-256 Hashing with Node.js <code>crypto</code> Module

Concept Overview


The <code>crypto</code> module is a built-in Node.js module that provides cryptographic functionality including hashing, encryption, and more. SHA-256 (Secure Hash Algorithm 256-bit) produces a fixed 64-character hexadecimal digest from any input string — it is one-way (cannot be reversed).

Full Implementation
<pre1.
// Step 1: Import the built-in crypto module
const crypto = require('crypto');

// Step 2: Define the input string
const inputString = 'password123';

// Step 3: Create SHA-256 hash and output as hexadecimal
const hash = crypto
  .createHash('sha256')      // Specify the algorithm
  .update(inputString)       // Feed in the data to hash
  .digest('hex');            // Output format: hexadecimal string

// Step 4: Display the result
console.log(`Original String : ${inputString}`);
console.log(`SHA-256 Hash    : ${hash}`);
```

---

### Output
```

SHA-256 Hash    : ef92b778bafe771e89245b89ecbc08a44a4e166c06659911881f383d4473e94f
</pre>


This hash is deterministic — "password123" will always produce this exact same hash.

2. <code>createHash('sha256')</code>

Creates a <code>Hash</code> object configured to use the SHA-256 algorithm. Other supported algorithms include <code>'md5'</code>, <code>'sha1'</code>, <code>'sha512'</code>, etc.

# Question 3: Use the crypto module to generate a secure, random Version 4 UUID.


Ans 3:

Concept Overview

A UUID (Universally Unique Identifier) is a 128-bit identifier standardized as a 36-character string in the format <code>xxxxxxxx-xxxx-xxxx-xxxx-xxxxxxxxxxxx</code>. A Version 4 UUID is entirely randomly generated (except for 4 version bits and 2 variant bits), making collisions astronomically unlikely. Node.js <code>crypto</code> module provides a built-in method to generate these securely.


Full Implementation
<pre>
// Step 1: Import the built-in crypto module
const crypto = require('crypto');

// Step 2: Generate a Version 4 UUID using randomUUID()
const uuid = crypto.randomUUID();

// Step 3: Display the result
console.log(`Generated UUID : ${uuid}`);
console.log(`UUID Length    : ${uuid.length} characters`);
console.log(`UUID Version   : ${uuid[14]}`);  // The 15th character is always '4'
```

---

### Output
```
Generated UUID : 3b12f1df-5232-4804-897e-917bf397618a
UUID Length    : 36 characters
UUID Version   : 4
</pre>

Every call to <code>randomUUID()</code> produces a different, cryptographically secure UUID.


1. Import <code>crypto</code>

- Built into Node.js (v14.17+ for <code>randomUUID())</code>. No installation required.

2. <code>crypto.randomUUID()</code>
<pre>
const uuid = crypto.randomUUID();
```
- Internally generates **16 cryptographically random bytes** using the OS-level CSPRNG (same source as `crypto.randomBytes()`).
- Sets **version bits** (`4`) at position 14.
- Sets **variant bits** (`8`, `9`, `a`, or `b`) at position 19.
- Formats into the standard `8-4-4-4-12` hyphenated string.

---

### UUID Structure Explained
```

# Question 4: Write a Node.js script that reads a PORT environment variable. If it's not set, it should default to 3000.


Ans 4:

 Concept OverviewEnvironment variables are external configuration values injected into a Node.js process at runtime via the operating system or a <code>.env</code> file. They are accessed through the global <code>process.env</code> object. This pattern is fundamental to the 12-Factor App methodology — keeping configuration separate from code.

 Full Implementation
 <pre>
 // Step 1: Read the PORT environment variable from process.env
// Step 2: If not set, default to 3000 using the || operator
const PORT = process.env.PORT || 3000;

// Step 3: Display the result
console.log(`Server will run on PORT: ${PORT}`);
console.log(`Source: ${process.env.PORT ? 'Environment Variable' : 'Default Value'}`);
```

---

### Output — When PORT is NOT set
```
Server will run on PORT: 3000
Source: Default Value
```

### Output — When PORT IS set (e.g. PORT=8080)
```
Server will run on PORT: 8080
Source: Environment Variable
</pre>

How to Set the Environment Variable


On Linux/macOS (inline):
<pre>
PORT=8080 node app.js
</pre>

On Windows (Command Prompt):
<pre>
set PORT=8080 && node app.js
</pre>


Step-by-Step Breakdown

1. <code>process.env</code>
<pre>process.env.PORT</pre>

- <code>process</code> is a global Node.js object — no import needed.
- <code>process.env</code> is a plain object containing all environment variables as strings.
- If <code>PORT</code> is not set, <code>process.env.PORT</code> returns <code>undefined</code>.

2. The || Fallback Operator

<pre>
javascriptconst PORT = process.env.PORT || 3000;</pre>

- If <code>process.env.PORT</code> is <code>undefined</code>, <code>null</code>, <code>''</code>, or <code>0</code>, JavaScript evaluates it as falsy and falls back to <code>3000</code>.
- This is the most concise and widely-used pattern for defaults.

3. Nullish Coalescing Operator <code>??</code> (Modern Alternative)
<pre>
javascriptconst PORT = process.env.PORT ?? 3000;</pre>

- Only falls back to <code>3000</code> if the value is strictly <code>null</code> or <code>undefined</code>.
- Does not fall back for 0 or '' — more precise than ||.

<pre>
Expression           <code>process.env.PORT = undefined</code>        <code>process.env.PORT = '' </code>                  process.env.PORT = '0'

|| fallback          uses 3000 ✅                       uses 3000 ✅                             uses 3000 ⚠️
?? fallback          uses 3000 ✅                       uses '' ⚠️                               uses '0' ✅

</pre>
For PORT specifically, <code>||</code> is preferred since empty string and 0 are invalid port values.
</pre>
Key Concepts Summary

<pre>
Concept                         Detail

process.env                     Global object holding all environment variables as strings

Fallback with ||                Returns default if value is falsy (undefined, '', 0)

Fallback with ??                Returns default only if value is null or undefined

.env file                       Stores config locally; loaded via dotenv package

Why it matters                  Allows same code to run in dev, staging, and production with different configs

Best Practice                   Never hardcode ports or secrets — always use environment variables

SET email = 'new.email@example.com';
```

**Result of missing WHERE:**
```
+----+-----------------------+---------------------+
| id | email                 | created_at          |
+----+-----------------------+---------------------+
|  1 | new.email@example.com | 2026-03-12 10:45:00 |
|  2 | new.email@example.com | 2026-03-12 10:45:01 |
|  3 | new.email@example.com | 2026-03-12 10:45:02 |
|  4 | new.email@example.com | 2026-03-12 10:45:03 |
+----+-----------------------+---------------------+
</pre>


Every user now has the same email — data is corrupted. Always include WHERE.

Safe Update Practice — Preview Before Updating

Always SELECT first to confirm which rows will be affected before running UPDATE:
<pre>
sql-- Step 1: Preview the row that will be affected
SELECT id, email FROM users
WHERE id = 1;

-- Step 2: Once confirmed, run the UPDATE
UPDATE users
SET email = 'new.email@example.com'
WHERE id = 1;

-- Step 3: Verify the change
SELECT id, email FROM users
WHERE id = 1;
</pre>

# Question 10: Given a users table and a posts table (with a user_id foreign key), write a query to retrieve all posts and the email of the user who wrote them.


Ans 10:

SQL JOIN Query — Retrieving Posts with Author Email
Concept Overview

A JOIN clause combines rows from two or more tables based on a related column between them. When data is split across multiple tables (like users and their posts), JOIN is how you reunite that data into a single meaningful result. This is the foundation of relational database design.


Reference: Table Structures

users table:
<pre>
sqlCREATE TABLE users (
  id         INT          PRIMARY KEY AUTO_INCREMENT,
  email      VARCHAR(255) NOT NULL UNIQUE,
  created_at DATETIME     DEFAULT CURRENT_TIMESTAMP
);
</pre>

posts table:
<pre>
CREATE TABLE posts (
  id         INT          PRIMARY KEY AUTO_INCREMENT,
  title      VARCHAR(255) NOT NULL,
  body       TEXT,
  user_id    INT          NOT NULL,
  created_at DATETIME     DEFAULT CURRENT_TIMESTAMP,
  FOREIGN KEY (user_id) REFERENCES users(id)
);
```

> `user_id` in `posts` is a **Foreign Key** — it references the `id` column in `users`, linking each post to its author.

---

### Sample Data

**`users`:**
```

Full Implementation
<pre>
-- Retrieve all posts along with the email of the user who wrote them
SELECT
  posts.id         AS post_id,
  posts.title      AS post_title,
  users.email      AS author_email
FROM posts
INNER JOIN users
  ON posts.user_id = users.id;
```

---

### Output
```

Types of JOINs Compared

<pre>
-- INNER JOIN: Only posts WITH a matching user (excludes orphan posts)
SELECT posts.id, posts.title, users.email
FROM posts
INNER JOIN users ON posts.user_id = users.id;

-- LEFT JOIN: ALL posts, even if no matching user exists
SELECT posts.id, posts.title, users.email
FROM posts
LEFT JOIN users ON posts.user_id = users.id;

-- RIGHT JOIN: ALL users, even if they have no posts
SELECT posts.id, posts.title, users.email
FROM posts
RIGHT JOIN users ON posts.user_id = users.id;
</pre>
